## Ejercicio 10: ¿Qué hay dentro de un Pipeline?

Un pipeline hace automáticamente estos pasos:

```
Entrada → Tokenizer → Modelo → Postprocesado → Salida
```

1. **Preprocesado (Tokenizer)**: Convierte el texto en tokens
2. **Modelo**: Realiza la inferencia y devuelve logits/probabilidades
3. **Postprocesado**: Convierte la salida numérica en etiquetas legibles

En este ejercicio vamos a **descomponer el pipeline** y usar cada componente por separado.

### Ejemplo: Clasificación de sentimiento "a mano"

Veamos cómo funciona internamente un pipeline de `sentiment-analysis`:

````python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

#Usamos el mismo modelo que usa el pipeline de sentiment-analysis por defecto
modelo_nombre = "distilbert-base-uncased-finetuned-sst-2-english"

# ========================================
# PASO 1: Cargar el Tokenizer y el Modelo
# ========================================
tokenizer = AutoTokenizer.from_pretrained(modelo_nombre)
modelo = AutoModelForSequenceClassification.from_pretrained(modelo_nombre)

texto = "I love this movie, it's amazing!"


# ========================================
# PASO 2: Tokenizar el texto
# ========================================
tokens = tokenizer(texto, return_tensors="pt")

print("=== PASO 2: TOKENIZACIÓN ===")
print(f"Texto original: {texto}")
print(f"Tokens (IDs): {tokens['input_ids']}")
print(f"Tokens decodificados: {tokenizer.convert_ids_to_tokens(tokens['input_ids'][0])}")
print()
````

````python
# ========================================
# PASO 3: Hacer inferencia (obtener los logits)
# ========================================

#Pasar los tokens por el modelo (inferencia)
with torch.no_grad():  #No necesitamos gradientes para inferencia
    outputs = modelo(**tokens)
    logits = outputs.logits

print("=== PASO 3: INFERENCIA DEL MODELO ===")
print(f"Logits (salida sin procesar del modelo): {logits}")
print(f"Shape de los logits: {logits.shape}")
print()
````

````python
# ========================================
# PASO 4: Postprocesado
# ========================================
#Convertir logits a probabilidades usando softmax
probabilidades = torch.nn.functional.softmax(logits, dim=-1)

#Obtener la clase predicha
clase_predicha = torch.argmax(probabilidades, dim=-1).item()

#Mapear el índice a la etiqueta
etiquetas = modelo.config.id2label

#Obtener la etiqueta correspondiente a la clase predicha
etiqueta = etiquetas[clase_predicha]

#Obtener el score (probabilidad) de la clase predicha
score = probabilidades[0][clase_predicha].item()

print("=== PASO 4: POSTPROCESADO ===")
print(f"Probabilidades: {probabilidades}")
print(f"Etiquetas disponibles: {etiquetas}")
print(f"\n>>> Resultado final: {etiqueta} (score: {score:.4f})")
````

### Comparación con el Pipeline

Veamos que obtenemos el mismo resultado:

````python
#Comparar con el pipeline
from transformers import pipeline

clasificador = pipeline("sentiment-analysis", model=modelo_nombre)
resultado_pipeline = clasificador(texto)

print("=== COMPARACIÓN ===")
print(f"Nuestro resultado manual: {etiqueta} ({score:.4f})")
print(f"Resultado del pipeline:   {resultado_pipeline[0]['label']} ({resultado_pipeline[0]['score']:.4f})")
print("\n¡Son iguales! El pipeline hace todo esto automáticamente.")
````

## Ejercicio 10.1: Configuración de inferencia en pipelines

En este ejercicio vais a practicar cómo controlar la salida de un pipeline con parámetros de inferencia.

### Parámetros importantes en pipelines
- `top_k`: cuántas predicciones devuelve el modelo ordenadas por confianza.
  - En modelos de **clasificación** (por ejemplo `text-classification`, `image-classification`, `zero-shot-classification`), el modelo calcula una puntuación para cada etiqueta posible.
  - Con `top_k=1` ves solo la etiqueta más probable.
  - Con `top_k=3` ves las 3 más probables, útil para analizar casos ambiguos.
- `truncation=True`: recorta entradas largas para evitar errores de longitud.
- `max_length`: fija la longitud máxima usada al truncar o al generar texto.
- `return_all_scores=True`: devuelve todas las clases con su score (en algunas versiones se usa `top_k=None`).

### Ejemplo de sintaxis
```python
text_classifier = pipeline("text-classification")
resultado = text_classifier("I love this", top_k=3)
```
